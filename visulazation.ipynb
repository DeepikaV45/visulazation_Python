```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=np.array([2,4,6,8])
plt.plot(x,y,marker='+',linestyle='--',color='b')
plt.title("Simple Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,2,3,4])
y=np.array([2,4,6,8])
x1=np.array([3,0,5,2])
y1=np.array([2,5,2,3])
plt.plot(x,y,marker='+',linestyle='--',color='b')
plt.plot(x1,y1,marker='o',linestyle='--',color='r')
plt.title("Simple Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['Apple','Banana','Cherry','Mango'])
y=np.array([10,20,45,23])
plt.bar(x,y,color='darkblue')
plt.title("Fruits Sales")
plt.xlabel("Fruits")
plt.ylabel("Quantity sold")
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['10','20','30','40'])
y=np.array([95,20,45,86])
plt.scatter(x,y,color='orange')
plt.title("Height vs Score")
plt.xlabel("Height")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
labels=['python','java','c++','js']
sizes=[20,40,15,20]
colors=['lightgreen','gold','lightblue','lightpink']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=140)
plt.title("Programming Languages")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('HistoricalPrices.csv')
df['Data']=pd.to_datetime(df['Date'])
df.sort_values('Date',inplace=True)
#line plot of closing prices
plt.figure(figsize=(12,6))
plt.plot(df['Date'],df[' Close'],color='blue',linestyle='--',marker='*')
plt.title("Closing prize Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Prize")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data={
    'product':['A','B','C','D'],
     'sales':[100,150,120,180]

}
df=pd.DataFrame(data)
sns.lineplot(x='product',y='sales',data=df)
plt.title("Sales Over Time")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("HistoricalPrices.csv")
df.columns.str.strip().str.upper()
num_df=df.select_dtypes(include='number')
corr_matrix=num_df.corr()
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm",linewidths="0.9")
plt.title("Corraltion - Heat Map")
plt.show()
```
